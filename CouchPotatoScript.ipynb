Script that creates the stimuli for CouchPotato Experiment into two files: english_stims.txt and viet_stims.txt

```python
import pandas as pd
```

```python
df = pd.read_csv('couch_potato_stims.csv')
```

```python
df
```

```python
adj_classes = [df["Class"].value_counts().index.tolist(), df["Class2"].value_counts().index.tolist()]
adj_classes = [adj_class for adj_col in adj_classes for adj_class in adj_col] # Unflatten list of lists

possible_adj_phrases = []

for adj_class in adj_classes:
    #print(adj_class)
    
    # Series for this adjective class
    series_adj1 = df.loc[df["Class"] == adj_class, "Adjective"]
    
    # If the Series was not located in first Class, check second Class
    if(len(series_adj1) == 0):
        series_adj1 = df.loc[df["Class2"] == adj_class, "Adjective2"]
    
    # List of all other adjective classes 
    classes_adj2 = [x for x in adj_classes if x != adj_class]
    #print(classes_adj2)
    
    for adj1 in series_adj1: # Loop through first adjective class
        for adj_class2 in classes_adj2: # Loop through every other class
            # Get Series for other adjective class
            series_adj2 = df.loc[df["Class"] == adj_class2, "Adjective"]
            
            if(len(series_adj2) == 0):
                series_adj2 = df.loc[df["Class2"] == adj_class2, "Adjective2"]
                
            for adj2 in series_adj2: # Loop through each adjective in other class
                #print('\t', adj1 + ' ' + adj2)
                
                possible_adj_phrases.append(adj1 + ' ' + adj2)


stims = []

# For each of the possible phrases, append each Noun to it
for noun in df["Noun"]:
    # Check if noun is not NaN by checking if it equals itself
    if(noun == noun):
        for adj_phrase in possible_adj_phrases:
            stims.append(adj_phrase + ' ' + noun)
```

```python
len(stims)
```

```python
stims
```

```python
with open('english_stims.txt', 'w') as f:
    for item in stims:
        f.write("%s\n" % item)
```

```python
# Do the same but in Vietnamese
viet_possible_adj_phrases = []

for adj_class in adj_classes:
    # Series for this adjective class
    series_adj1 = df.loc[df["Class"] == adj_class, "Tính từ"]
    
    # If the Series was not located in first Class, check second Class
    if(len(series_adj1) == 0):
        series_adj1 = df.loc[df["Class2"] == adj_class, "Tính từ2"]
    
    # List of all other adjective classes 
    classes_adj2 = [x for x in adj_classes if x != adj_class]
    #print(classes_adj2)
    
    for adj1 in series_adj1: # Loop through first adjective class
        for adj_class2 in classes_adj2: # Loop through every other class
            # Get Series for other adjective class
            series_adj2 = df.loc[df["Class"] == adj_class2, "Tính từ"]
            
            if(len(series_adj2) == 0):
                series_adj2 = df.loc[df["Class2"] == adj_class2, "Tính từ2"]
                
            for adj2 in series_adj2: # Loop through each adjective in other class
                #print('\t', adj1 + ' ' + adj2)
                
                viet_possible_adj_phrases.append(adj1 + ' ' + adj2)


viet_stims = []

# For each of the possible phrases, append each Noun to it
for noun in df["Danh từ"]:
    # Check if noun is not NaN by checking if it equals itself
    if(noun == noun):
        for adj_phrase in viet_possible_adj_phrases:
            viet_stims.append(noun + ' ' + adj_phrase)
```

```python
len(viet_stims)
```

```python
with open('viet_stims.txt', 'w', encoding = 'utf-8') as f:
    for item in viet_stims:
        f.write("%s\n" % item)
```
